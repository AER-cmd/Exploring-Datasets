```python
from astroquery.vizier import Vizier
from astropy.coordinates import  SkyCoord
import astropy.units as u

targets = []
for i in range(20): 

    coord = SkyCoord(ra=ra_value[i]*u.deg, dec=dec_value[i]*u.deg, frame='icrs') 
    vizier = Vizier(columns=["*", "_r"], row_limit=10)  
    vizier.VIZIER_SERVER = "vizier.cds.unistra.fr"
    result = vizier.query_region(coord, radius=5*u.arcsec)
    targets.append(result)
    print(result.keys())
    

```

```python
for i in range(len(targets)):
    print(targets[i].keys())
```

```python
list_dict = []
for i in range(len(targets)):
    dict = {"cat1": targets[i]['I/359/ssoobs']['Name']}
    list_dict.append(dict)
print(list(list_dict))  
```
